```python
from bs4 import BeautifulSoup #digunakan untuk web scraping
import requests #digunakan untuk mendapatkan/mengakses/mendownload Webpage
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

```python
#url of the page we want to scrape
# url = "https://namelix.com/app/?keywords=data%2C+software%2C+science"1
# url = "https://namelix.com/app/?keywords=future%2C+data%2C+metaverse%2C+ai"
url = "https://namelix.com/app/?keywords=future%2C+data+scientist%2C+machine+learning%2C+ai%2C+space"



# initiating the webdriver. Parameter includes the path of the webdriver.
driver = webdriver.Firefox()
driver.get(url) 
  
# this is just to ensure that the page is loaded
time.sleep(20) 

elem = driver.find_element(by=By.TAG_NAME, value='body')

no_of_pagedowns = 1000

while no_of_pagedowns:
    elem.send_keys(Keys.PAGE_DOWN)
    time.sleep(1)
    no_of_pagedowns-=1  
    
html = driver.page_source
```

```python
# this renders the JS code and stores all
# of the information in static HTML code.
domain_df = pd.DataFrame()
domain_list = []
# Now, we could simply apply bs4 to html variable
soup = BeautifulSoup(html, "html.parser")
#print(soup)
logowrappers = soup.find_all('div', {'class' : 'logo-wrapper'})
for lw in logowrappers:
  logopremium = lw.find_all('div', {'class' : 'logo premium'})
  #print("1")
  if(logopremium != []):
    continue 
  #print("2") 
  title = lw.find_all('g', {'id' : 'title'})  
  #print(title)
  if(title == []):
    continue
  domain = title[0]['aria-label']
  dec = lw.find('span', {'class' : 'description'})
  #print(dec.text)
  avail = lw.find('span', {'class':'btn btn-sm btn-rounded-circle btn-light border-white bg-white'})
  #print(avail)
  if(avail is None) :
    ada = "N/A"
  else:
    ada = "OK"
    dict = {"domain":domain, "availabilty":ada, "description":dec.text}
    domain_list.append(dict)

driver.close() # closing the webdriver
domain_df = pd.DataFrame(domain_list)
domain_df.to_excel("future data scientist machine learning ai space 1000page down - medium randomness two words.xlsx") 
```
